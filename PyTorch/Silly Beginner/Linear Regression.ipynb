```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Variable
```

```python
x = np.linspace(-1, 1, 100).reshape((100, 1))
y = 2 * x + 3 + np.random.randn(100, 1)*0.05
```

```python
model = nn.Linear(1,1)
loss = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.5)
```

```python
loss_list = []
m_list = []
b_list = []
for _ in range(20):
    output = model(Variable(torch.from_numpy(x.astype('float32'))))
    optimizer.zero_grad()
    
    l = loss(output, Variable(torch.from_numpy(y.astype('float32'))))
    l.backward()
    optimizer.step()
    
    loss_list.append(l.data.numpy())
    m_list.append(list(model.parameters())[0][0].data.numpy()[0])
    b_list.append(list(model.parameters())[1][0].data.numpy()[0])
```
